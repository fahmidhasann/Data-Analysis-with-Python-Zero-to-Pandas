Let's merge the two data frames, and compute some more metrics.

**Q9: Merge `countries_df` with `covid_data_df` on the `location` column.**

*Hint: Use the `.merge` method on `countries_df`.

```python
combined_df = covid_data_df.merge(countries_df,on='location')
```

```python
combined_df
```

**Q10: Add columns `tests_per_million`, `cases_per_million` and `deaths_per_million` into `combined_df`.**

```python
combined_df['cases_per_million']= (combined_df.total_cases/combined_df.population)*1e6
```

```python
combined_df['tests_per_million'] = (combined_df.total_tests/combined_df.population)*1e6
```

```python
combined_df['deaths_per_million'] = (combined_df.total_deaths/combined_df.population)*1e6
```

**Q11: Create a dataframe with 10 countires that have highest number of tests per million people.**

```python
ten_highest_tests_df = combined_df.sort_values('tests_per_million',ascending=False)[['location','tests_per_million']].head(10)
```

```python
ten_highest_tests_df
```

**Q12: Create a dataframe with 10 countires that have highest number of positive cases per million people.**

```python
ten_highest_cases_df = combined_df.sort_values('cases_per_million',ascending=False)[['location','cases_per_million']].head(10)
```

```python
ten_highest_cases_df
```

**Q13: Create a dataframe with 10 countires that have highest number of deaths cases per million people?**

```python
ten_highest_deaths_df = combined_df.sort_values('deaths_per_million',ascending=False)[['location','deaths_per_million']].head(10)
```

```python
ten_highest_deaths_df
```

**(Optional) Q: Count number of countries that feature in both the lists of "highest number of tests per million" and "highest number of cases per million".**

```python
common_countries1=ten_highest_cases_df.merge(ten_highest_deaths_df,on='location',how='inner')
common_countries1.location.count()
```
